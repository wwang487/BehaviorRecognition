The following cells are designed to test the accuracy of your model.

The first step is to extract all the images and markings in the testing dataset, you can do this by running **python ExtractTestImgs.py** in the same terminal as above training steps. After running this, you should say a folder called Accuracy_Test being created in your project folder, and within this folder, there should be two sub-folders, one is GT_Marking saved your marking for the testing images, and one is Imgs saved the testing images.

The second step is to test performance of the trained model in different epochs for the testing dataset by comparing the model detection results with the manual markings. You need first run **TestMarkingResDemo.py**, this code will generate the detection results using the models saved in each training epoch, an example function call is:

**python TestMarkingResDemo.py --net res101 --startepoch 1 --endepoch 200 --checkpoint 4298 --cuda --load_dir models**

Please first modify lines 185-191 to change the classes the same as your markings, then make sure the load_dir is the same as where you save your trained model, the startepoch is the same as the smallest epoch you saved, the endepoch is the largest epoch you saved, and the checkpoint is the same as your training dataset size. If your model is named as: faster_rcnn_1_50_1892.pth, then epoch is 50, and size is 1892. Once you get the values for **--startepoch --endepoch --checkpoint** and **--load_dir** fixed, you can copy the modified function call to the terminal, and run it.

Then, you need to run **NewTestBetweenDetectionAndGT1.py** to compare the detection results with your markings. This function has a lot of arguments.You need to set **--box_folder** the same folder you saved your detection txt file, it should be saved at **./Accuracy_Test/boxes/** or **./Accuracy_Test/boxes/Imgs/**, set **--gt_folder** the same as the folder you saved your manual marking results for the testing dataset, it should be saved at **./Accuracy_Test/GT_Marking/**, set the **--startepoch** and **--endepoch** the same as above. The following is a sample function call:

**python NewTestBetweenDetectionAndGT1.py --startepoch 1 --endepoch 200 --box_folder ./Accuracy_Test/boxes/Imgs/ --gt_folder ./Accuracy_Test/GT_Marking/**

You can modify the call with your settings and copy-paste the modified command into the terminal, and run it.

After running, all of your testing results should be saved into **./Accuracy_Test/tables/**
